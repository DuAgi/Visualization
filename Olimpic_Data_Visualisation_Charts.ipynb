#Import libraries and import data

```python
# import 'pandas' data analysis library
import pandas as pd

# import 'matplotlib' library's 'pyplot' interface for creating charts/plots
import matplotlib.pyplot as plt

# import chart style sheets
from matplotlib import style
# which style sheet to use
style.use('ggplot')
```

```python
inputfolderpath = r'C:\Users\dunca_n0wxpzg\Documents\Hobbi projekt\Tableau'
inputfile = 'olimpia_1.csv'
```

```python
# read data from csv file into a dataframe
df = pd.read_csv(inputfolderpath + '\\' + inputfile, sep = ',', encoding = 'utf-8')
df.head()
```

```python
df.shape
```

```python
# show data types of all columns
print(df.dtypes)
```

```python
# create a pivot table
pivot_sum_gold_year = pd.pivot_table(df, values = ['Gold_dummy'], index = ['Year'], 
    aggfunc = 'sum', fill_value = 0)
pivot_sum_gold_year
```

```python
# sort the table
pivot_sum_gold_year = pivot_sum_gold_year.sort_values(by = ['Gold_dummy'], ascending = False)
pivot_sum_gold_year
```

```python
# create a pivot table
pivot_sum_gold_country_year = pd.pivot_table(df, values = ['Gold_dummy'], index = ['Year'], columns = ['Country'], 
    aggfunc = 'sum', fill_value = 0, margins = True, margins_name = 'Total')
pivot_sum_gold_country_year
```

```python
# define a function to set chart properties
def set_chart_properties(xlabel, ylabel, title, legendlist, legendloc, xtickrotation):
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.legend(legendlist, loc = legendloc)
    plt.xticks(rotation = xtickrotation, ha = 'right')
    plt.show()
```

#Data visualizations, charts

```python
x = pivot_sum_gold_country.index
y = pivot_sum_gold_country['Gold_dummy']
```

```python
# create a figure
plt.figure()

# default line chart
plt.plot(x, y)

plt.xlabel('Country')
plt.ylabel('Gold_dummy')
plt.title('Gold medals')
plt.legend(['gold medals'], loc = 'upper right')
plt.xticks(rotation = 45, ha = 'right')
plt.show()
```

```python
# create a figure
plt.figure()

# line chart (blue solid line)
plt.plot(x, y, 'b-')

set_chart_properties('Country', 'Gold_dummy', 'Gold medals', ['gold medals'], 'upper right', 45)
```

```python
# create a figure
plt.figure()

# line chart (yellow, circle markers, solid line)
plt.plot(x, y, 'yo-')

# area chart (red circle markers, fill in orange)
plt.plot(x, y, 'ro')
plt.fill_between(x, y, color = 'orange')



set_chart_properties('Country', 'Gold_dummy', 'Gold medals', ['gold medals'], 'upper right', 45)
```

```python
# create a figure
plt.figure()

# vertical bar (column) chart
plt.bar(x, y)


set_chart_properties('Country', 'Gold_dummy', 'Gold medals', ['gold medals'], 'upper right', 45)
```

```python
# create a figure
plt.figure()

# vertical bar (column) chart
plt.bar(x, y, width = 0.5, color = 'skyblue', edgecolor = 'darkblue')

set_chart_properties('Country', 'Gold_dummy', 'Gold medals', ['gold medals'], 'upper right', 45)
```

```python
# create a figure
plt.figure()

# horizontal bar chart
plt.barh(x, y, height = 0.5, color = 'skyblue', edgecolor = 'darkblue')

# reverse the values on y axis
plt.gca().invert_yaxis()


set_chart_properties('Country', 'Gold_dummy', 'Gold medals', ['gold medals'], 'upper right', 45)
```

```python
# create a figure
plt.figure()

# pie chart
plt.pie(y, labels = x)

plt.title('Gold medals')
plt.show()
```

#Multiple charts

```python
# create a pivot table
pivot_sum_gold_year_gender = pd.pivot_table(df, values = ['Gold_dummy'], index = ['Year'], columns = ['Gender'], aggfunc = 'sum', fill_value = 0)
pivot_sum_gold_year_gender
```

```python
# remove multiple column header
pivot_sum_gold_year_gender = pivot_sum_gold_year_gender['Gold_dummy']
pivot_sum_gold_year_gender
```

```python
# create a figure
plt.figure()

# multiple line chart
pivot_sum_gold_year_gender.plot.line(color = ['tomato', 'lightseagreen'])
plt.xticks(rotation = 45, ha = 'right')
plt.title('Gold by Gender and Year')
plt.show()
```

```python
x = pivot_sum_gold_year_gender.index
y1 = pivot_sum_gold_year_gender['Men']
y2 = pivot_sum_gold_year_gender['Women']

```

```python
# create a figure
plt.figure()

# multiple line chart
plt.plot(x, y1, 'yo-')
plt.plot(x, y2, 'bo-')

set_chart_properties('Year', 'Gold medal', 'Gold by Gender and Year', 
    ['Men', 'Women'], 'upper right', 45)
```

```python
# create a figure
plt.figure()

# stacked vertical bar (column) chart
pivot_sum_gold_year_gender.plot.bar(stacked = True, 
    color = ['tomato','lightseagreen'])
plt.title('Gold by Gender and Year')
plt.show()
```

```python
# create a figure
plt.figure()

# stacked horizontal bar chart
pivot_sum_gold_year_gender.plot.barh(stacked = True, 
    color = ['tomato','lightseagreen'])
plt.title('Gold by Gender and Year')
plt.show()
```
